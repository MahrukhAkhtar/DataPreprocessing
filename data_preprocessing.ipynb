### importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### importing the dataset

```python
dataset = pd.read_csv("Data.csv")
dataset
X = dataset.iloc[:,0:3].values #need only column values
y = dataset.iloc[:,3].values
```

### taking care of missing data

```python
X
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean') #strategy means onhe means se badal do
imputer.fit(X[:,1:3]) #imputer columns ka mean nikale ga
X[:,1:3] = imputer.transform(X[:,1:3]) 
X
```

### Ecoding  categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[0])], remainder='passthrough')
X = np.array(ct.fit_transform(X)) # change categorical into numerical
```

```python
X
```

### Encoding the dependent variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

### Spliting the dataset into training and test set

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.2, random_state=32)
X_train
```

```python
y_test
```

### feature scalling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```
